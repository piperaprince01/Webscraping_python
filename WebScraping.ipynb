# Importing the Libraries

```python
import numpy as np
import pandas as pd
import requests
import json
```

# Web Scraping

As it is a dynamically generated data, I had to view the network section of the inspect element to come up with this link for the json data

```python
url = 'https://techolution.app.param.ai/api/career/get_job/?query=&locations=&category=&job_types='
```

```python
response = requests.get(url)
```

```python
print(response)
```

The above code shows that the response worked as it is 200

```python
response.text
```

The above text/output shows the detailed data that was needed by us and the uRL provided 

```python
data = response.text
```

Here we will do json parsing

```python
parsed = json.loads(data)
```

```python
parsed
```

```python
joddetails = parsed["data"]
```

```python
joddetails
```

```python
Consultingjobs = joddetails["Consulting"]
```

```python
Consultingjobs
```

```python
ITJobs = joddetails["Information Technology"]
```

```python
FRjobs = joddetails["Facial recognition"]
```

Importing json normalize to convert json data to pandas Dataframe

```python
from pandas.io.json import json_normalize
```

Converting each category individually

```python
IT = json_normalize(ITJobs['jobs'])
```

```python
Consulting = json_normalize(Consultingjobs['jobs'])
```

```python
FR = json_normalize(FRjobs['jobs'])
```

```python
Consulting
```

```python
IT.head()
```

Adding all 3 dataframes to form one Dataframe

```python
df_jobs = [Consulting,IT,FR]
```

```python
df_finaljobs = pd.concat(df_jobs)
```

```python
df_finaljobs.head()
```

```python
df_finaljobs = df_finaljobs.reset_index()
```

```python
df_finaljobs.shape
```

Sorting the dataframe as per created the starting with the oldest first

```python
df_finaljobs = df_finaljobs.sort_values(by='created_at')
```

```python
df_finaljobs.head()
```

Code to extract the dataframe to csv

```python
df_finaljobs.to_csv('Technolution_careers.csv')
```
